```python
import bs4 as bs
from urllib import request
```

```python
sauce = request.urlopen('https://en.wikipedia.org/wiki/William_Shakespeare').read()
```

```python
sauce
```

```python
soup = bs.BeautifulSoup(sauce,'lxml')
```
