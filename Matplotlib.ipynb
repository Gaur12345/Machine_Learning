```python
x1=[1,2,3]
y1=[5,7,4]
x2=[1,2,3]
y2=[10,14,12]

plt.plot(x1,y1,label="First Plot",color='red')
plt.plot(x2,y2,label="First Plot",color='green')

plt.xlabel('X ka Label')
plt.ylabel('Y ka Label')
plt.title("Ye mera Graph hai")
plt.legend()

```

```python
x=[1,3,5,7,9]
y=[5,2,6,3,6]

xx=[2,4,6,8]
yx=[4,5,6,7]

plt.bar(x,y,label='Pehla plot',color='red')
plt.bar(xx,yx,label='dusra plot',color='pink')

plt.title('Bar chart')
plt.xlabel('x Asix')
plt.ylabel('y Axis')

plt.legend()
```

```python
x=[1,2,3,4,5,6]
y=[4,3,5,3,3,6]

plt.scatter(x,y,label='scatter',color='red',marker='<',s=202)
plt.legend()
```

```python
slices = [95,34,32,35,12]
channels=['FTV','9xm','Ten Sports','Sony','Hindi Coding Zone']
cols=['red','green','cyan','pink','yellow']
plt.pie(slices,labels=channels,colors=cols,explode=(0.05,0,0,0,0),shadow=True,autopct='%1.1f%%',startangle=120)
plt.title('Pie chart')
```
