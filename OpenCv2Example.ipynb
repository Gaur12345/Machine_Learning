```python
# Image Translation 
img = cv2.imread('D:\Photos\Saved Pictures\slider11.jpg')
cv2.imshow('Orignal Image',img)
#  Store height and width of the image
height,width = img.shape[:2]
print(height)
print(width)

quarter_height ,quarter_width = height/4 ,width/4
print(quarter_height)
print(quarter_width)

#     [1 0 Tx]
# T =[o  1 Ty]

# T is Our Translation Matrix
T = np.float32([[1,0,quarter_width],
               [0,1,quarter_height]])

print(T)

# We use warpAffine Transformation to shift the image
img_translation = cv2.warpAffine(img, T,(width,height))

cv2.imshow('Tranlation Image',img_translation)

#Save the image
cv2.imwrite('E:\openCvImage\Translation Image.jpg',img_translation)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Image Rotation

img = cv2.imread('D:\Photos\Saved Pictures\slider11.jpg')

height,width = img.shape[:2]
rotation_matrix = cv2.getRotationMatrix2D((width/2,height/2),120,.3)
rotated_image =cv2.warpAffine(img,rotation_matrix,(width,height))

cv2.imshow('Origal image',img)
cv2.imshow('Roteted Image',rotated_image)
cv2.imwrite('E:\openCvImage\Rotation Image.jpg',rotated_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Image Transpose
img=cv2.imread('D:\Photos\Saved Pictures\slider11.jpg')

rotated_img = cv2.transpose(img)

roo=cv2.transpose(rotated_img)

cv2.imshow('orignal Image',img)
cv2.imshow('Rotated Image',rotated_img)
cv2.imshow("Last Image ",roo)
cv2.imwrite('E:\openCvImage\Rotated Image.jpg',rotated_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Scaling or resizing and interplotation
img=cv2.imread('D:\Photos\Saved Pictures\slider11.jpg')
cv2.imshow('orignal Image',img)

#Let's mok the size of our image 3/4 of it's orignal size
img_scaled = cv2.resize(img,None,fx=0.75,fy=0.75)
cv2.imshow('Scaling - Linear Interplotation',img_scaled)

#Let's doubled the size of our image
img_double=cv2.resize(img,None,fx=1.5,fy=1.5,interpolation=cv2.INTER_CUBIC)
cv2.imshow('Scaling - cubic Interpolation image ',img_double)

#Let's skew the re-sizing by setting exact dimensions/user define dimension
img_skew=cv2.resize(img,(900,400),interpolation=cv2.INTER_AREA)
cv2.imshow('Skew Image',img_skew)

# save all the image
cv2.imwrite('E:\openCvImage\Scaling - Linear  Image.jpg',img_scaled)
cv2.imwrite('E:\openCvImage\Scaling - Cubic  Image.jpg',img_double)
cv2.imwrite('E:\openCvImage\Scaling - Skew Image.jpg',img_skew)



cv2.waitKey()
cv2.destroyAllWindows()

```

```python
#Image resizing using pyrmid
img=cv2.imread('D:\Photos\Saved Pictures\slider11.jpg')


smaller=cv2.pyrDown(img)
larger=cv2.pyrUp(img)


cv2.imshow('orignal Image',img)
cv2.imshow('smaller image',smaller)
cv2.imshow('larger image',larger)

# save all the image
cv2.imwrite('E:\openCvImage\Smaller Image.jpg',smaller)
cv2.imwrite('E:\openCvImage\larger Image.jpg',larger)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# image cropping

img=cv2.imread('D:\Photos\Saved Pictures\slider11.jpg')

height,width = img.shape[:2]

#Let's get the starting pixel coordinate(top left,of cropping rectangle)
start_row,start_col = int(height*.25),int(width*.25)

#let's get the ending pixel coodinate (bottom right)
end_row,end_col = int(height*.75),int(width*.75)
#Simply use the indexing to crop the image

cropped =img[start_row:end_row,start_col:end_col]

cv2.imshow('orignal Image',img)
cv2.imshow('Cropped Image',cropped)

cv2.imwrite('E:\openCvImage\cropped Image.jpg',cropped)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img=cv2.imread('D:\Photos\Saved Pictures\slider11.jpg')
cv2.imshow('Orignal Image',img)

# Creating 3X3 Kernal

kernal_3x3 = np.ones((3,3),np.float32)/9

#We use the cv2.filter2D to convole the kernal with on image
blurred = cv2.filter2D(img,-1,kernal_3x3)
cv2.imshow('3x3 kernal Blurring',blurred)


#creating 7X7 Kernal
kernal_7x7 = np.ones((7,7),np.float)/49
blurred2=cv2.filter2D(img,-1,kernal_3x3)
cv2.imshow('7X7 Kernal Blurring',blurred2)

cv2.imwrite('E:\openCvImage\Blurring3X3 image.jpg',blurred)
cv2.imwrite('E:\openCvImage\Blurring7X7 Image.jpg',blurred2)
print(blurred)
cv2.waitKey()
cv2.destroyAllWindows()
```
