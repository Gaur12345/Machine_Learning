```python
from sklearn.tree import DecisionTreeClassifier
```

```python
kypo_predictor=DecisionTreeClassifier()
```

```python
x=data.drop("Kyphosis",axis=1)
```

```python
y=data.Kyphosis
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_train
```

```python
kypo_predictor.fit(x_train,y_train)
```

```python
pred=kypo_predictor.predict(x_test)
```

```python
pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,pred)
```

```python
kypo_predictor.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_kypo_predictor=RandomForestClassifier()
```

```python
random_kypo_predictor.fit(x_train,y_train)
```

```python
random_kypo_predictor.predict(x_test)
```

```python
random_kypo_predictor.score(x_test,y_test)
```
