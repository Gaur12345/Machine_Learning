```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv('E:\ML\Dataset\Social_Network_Ads.csv')
```

```python
data.head()
```

```python
real_x=data.iloc[:,[2,3]].values
real_x
```

```python
real_y = data.iloc[:,4].values
```

```python
real_y
```

```python
data.info()
```

```python
trainig_x,test_x,trainig_y,test_y = train_test_split(real_x,real_y,test_size=0.25,random_state=0)
```

```python
trainig_x
```

```python
test_x
```

```python
scaler = StandardScaler()
```

```python
trainig_x=scaler.fit_transform(trainig_x)
test_x=scaler.fit_transform(test_x)
```

```python
trainig_x
```

```python
classifi_LR = LogisticRegression(random_state=0)
classifi_LR.fit(trainig_x,trainig_y)
```

```python
y_pred = classifi_LR.predict(test_x)
```

```python
y_pred
```

```python
test_y
```

```python
c_m = confusion_matrix(test_y,y_pred)
```

```python
c_m
```

```python
from matplotlib.colors import ListedColormap
x_set, y_set = trainig_x,trainig_y
x1,x2 = np.meshgrid(np.arange(start=x_set[:,0].min()-1,stop=x_set[:,0].max()+1,step=0.01),
                    np.arange(start=x_set[:,1].min()-1,stop=x_set[:,1].max()+1,step=0.01))
plt.contourf(x1,x2,classifi_LR.predict(np.array([x1.ravel(),x2.ravel()]).T).reshape(x1.shape),
             alpha=0.75,cmap=ListedColormap(('red','green')))

plt.xlim(x1.min(),x1.max())
plt.ylim(x2.min(),x2.max())

for i , j in enumerate(np.unique(y_set)):
                    plt.scatter(x_set[y_set==j,0],x_set[y_set==j,1], 
                        c=ListedColormap(('red','green'))(i),label=j)

                    
plt.title('Logostic Regression (Training Set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
# Testing Dataset
x_set, y_set = test_x,test_y
x1,x2 = np.meshgrid(np.arange(start=x_set[:,0].min()-1,stop=x_set[:,0].max()+1,step=0.01),
                    np.arange(start=x_set[:,1].min()-1,stop=x_set[:,1].max()+1,step=0.01))
plt.contourf(x1,x2,classifi_LR.predict(np.array([x1.ravel(),x2.ravel()]).T).reshape(x1.shape),
             alpha=0.75,cmap=ListedColormap(('red','green')))

plt.xlim(x1.min(),x1.max())
plt.ylim(x2.min(),x2.max())

for i , j in enumerate(np.unique(y_set)):
                    plt.scatter(x_set[y_set==j,0],x_set[y_set==j,1], 
                        c=ListedColormap(('red','green'))(i),label=j)

                    
plt.title('Logostic Regression (Testing Set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
