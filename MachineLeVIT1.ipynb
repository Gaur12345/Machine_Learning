```python
#  Drow Line Plot
t = np.arange(0.0,2.0,0.01)
t
s = 1+np.cos(2*np.pi*t)
plt.plot(t,s,"*")
plt.xlabel("Time")
plt.ylabel("Voltage")
plt.title("cosine wave plot(Cos(x))")
plt.grid()
plt.show()
```

# subplot

```python
x1= np.linspace(0.0,5.0,45)
x2= np.linspace(0.0,2.0,45)
print("This is x1",x1)
print("Hello")
x2

```

```python
y1= np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)
plt.subplot(2,1,1)
plt.plot(x1,y1,'o-')
plt.title("Subplot 1")
plt.xlabel("x1")
plt.ylabel("Amp(y1)")

plt.subplot(2,1,2)
plt.plot(x2,y2,'.-')
plt.title('Subplot 2')
plt.xlabel('x2')
plt.ylabel('Amp(y2)')
plt.show()
```

```python
plt.plot([1,2,3])
# now create a subplot which represents the top plot of a grid
# with 2 rows and 1 column. Since this subplot will overlap the
# first, the plot (and its axes) previously created, will be removed
plt.subplot(211)
```

```python
plt.subplot(221)

# equivalent but more general
ax1=plt.subplot(2, 2, 1)

# add a subplot with no frame
ax2=plt.subplot(222, frameon=False)

# add a polar subplot
plt.subplot(223, projection='polar')

# add a red subplot that shares the x-axis with ax1
plt.subplot(224, sharex=ax1, facecolor='red')

# delete ax2 from the figure
plt.delaxes(ax2)

# add ax2 to the figure again
plt.subplot(ax2)
```

# Histogram plot

```python
ages = [20,22,24,26,31,43,13,35,12,67,68,21,68,98,14,14,13,76,56]
range=(0,100)
bins =10
plt.hist(ages,bins,range,color="Green",histtype='bar',rwidth=0.90)
plt.hist()
plt.xlabel('Ages')
plt.ylabel('Bins')
plt.title('Histogram Plot')
plt.show()
```

# Plot Merging

```python
x = np.arange(0,2*np.pi,0.01)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1,label='Sin')
plt.plot(x,y2,label='Cos')
plt.xlabel('X-axis')
plt.ylabel('Y=aixs')
plt.title('Sine & Cosine Function')
plt.legend()
plt.show()
```

```python
import csv
```

```python
x = []
y = []
with open('friend.csv','r') as csvfile:
    plots = csv.reader(csvfile)
    for col in plots:
        x.append(col[0])
        y.append(col[1])
        
        
plt.plot(x,y,label='File')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Reading Data')
plt.legend()
plt.show()
```
