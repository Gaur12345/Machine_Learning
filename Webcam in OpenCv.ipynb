```python
import cv2
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret,frame = cap.read()
    cv2.imshow('Our live Sketch',frame)
    if cv2.waitKey(1)==13:
        break
   


cap.release()
cv2.destroyAllWindows()
```
