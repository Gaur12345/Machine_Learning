# Matplotlib 

```python
import numpy as np
import matplotlib.pyplot as plt
import requests
import json
```

```python
plt.plot([1,2,3,3],[4,2,4,2])
plt.show()
```

```python
list1=[1,2,3,4]
list2=[5,6,7,8]
plt.plot(list1,list2)
plt.show()
```

```python
x1=[1,2,3]
y1=[5,7,4]
x2=[1,2,3]
y2=[10,14,12]

plt.plot(x1,y1,label="First plot",color='red')
plt.plot(x2,y2,label="Second Plot",color='green')

plt.xlabel('X Axis label')
plt.ylabel('y Axis label')
plt.title('ye mera ghar Hai')

plt.legend()
```

```python
x=[1,3,5,7,9]
y=[5,2,6,3,6]
xx=[2,4,6,8]
yy=[4,5,6,7]

plt.bar(x,y,label="Pehal Plot",color='red')
plt.bar(xx,yy,label="dusra plot",color="blue")
plt.title('Bar Chart')
plt.xlabel('x Axis')
plt.ylabel('Y Axis')
plt.legend()
```

```python
x=[1,2,3,4,5,6]
y=[4,3,5,3,3,6]
plt.scatter(x,y,label='Scatter Chart',color='red',marker='*',s=200)
#plt.show()
plt.legend()
```

```python
slices = [95,34,32,35,12]
channels = ['FTV','9xm','Ten Sports','Sony','Hindi codingzone']
cols = ['red','green','blue','cyan','yellow']
plt.pie(slices,labels=channels,colors=cols,explode=(0.05,0,0,0,0),shadow=True,autopct='%1.1f%%',startangle=120)
#plt.legend()
plt.show()
```
